```python
# Till now we saw how to plot graphs on static data given in the code, 
# now we will fetch the input coordinates from a text file 
# In the code shown below we fetch data from a file called example
 
import matplotlib.pyplot as plt
import csv                    #we use csv for Comma Seperated Values

x=[]     #create empty lists
y=[]


with open("/home/shweta/Desktop/example",'r') as csvfile:
    plots =csv.reader(csvfile,delimiter=',')    #this takes values from example separated by ',' in plots
    for row in plots:
        x.append(int(row[0]))     #append values to empty lists
        y.append(int(row[1]))

        print row        # this is optional and simply to print data of example file

plt.plot(x,y)

plt.title("Loaded data from file")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Now suppose we dont have numeric data on some axis 
# Lets take an example of a bar graph that shows percentage of students 
# Those percentage are stored in a file called example2.txt
 
import matplotlib.pyplot as plt
import csv                    #we use csv for Comma Seperated Values

x=[]     #create empty lists
y=[]


with open("Desktop/example2.txt",'r') as csvfile:
    plots =csv.reader(csvfile,delimiter=',')    #this takes values from example separated by ',' in plots
    for row in plots:
        x.append(row[0])     #append values to empty lists without integer conversion
        y.append(int(row[1]))

        print row        # this is optional and simply to print data of example file

plt.bar(x,y,width=0.5,color='c')

plt.title("Loaded data from file",color='r')
plt.xlabel("Students",color='b')
plt.ylabel("Percentage",color='b')
plt.show()
```
