```python
#python codes for different graph plots using matplotlib

#1. Let us first plot a Bar graph

import matplotlib.pyplot as plt

x=[2,4,6,8,10]
y=[12,5,8,17,20] #these are the x and y coordinate lists

a=[1,3,5,7,9]
b=[5,6,8,12,34] #we take another pair of coordinates just to differentiate multiple bars

plt.bar(x,y,label="bar1",color='red')  #bar() is function for bargraph
plt.bar(a,b,label="bar2",color='cyan') #label and color are attributes

plt.title("data 2018") #this is the title of graph
plt.xlabel("X-axis") #this way we provide labels to x and y axis
plt.ylabel("Y-axis") 
plt.legend()   #legend is used to display color representation as shown at the left top corner

plt.show()     #show() fuction is used to display the graph

```

```python
#2. Let us plot a histogram on age groups 

import matplotlib.pyplot as plt

#list of ages of different people
people_ages=[34,30,20,40,70,44,55,34,22,78,67,15,9,4,3,16,18,58,49,80,30,38,29,18,47,100,98,91,61]

bins=[0,10,20,30,40,50,70,80,90,100] # create histogram in ten years range i.e. create subgroups of 10 years each 

plt.hist(people_ages,bins,histtype="bar",rwidth=0.8,color='y') #hist() is function for histogram

plt.title("data 2018")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
#3. Let us plot a scatter plot for some coordinates

import matplotlib.pyplot as plt

x=[5,8,9,3,6,1,7,4,10,12,2]
y=[7,9,8,4,3,6,5,1,6,13,10]

#scatter() is the function to plot dots of said data and with marker parameter we can change the shape of dots being marked

plt.scatter(x,y,label='interesting data',color='b',marker='*',s=100) 

plt.title("data 2018")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# 4. Now we will plot a Stackplot 

import matplotlib.pyplot as plt

#We take a scenerio of recording our per day activities for a week

daysofweek=[1,2,3,4,5,6,7]

sleep=[7,8,6,7,7,8,9]  #these are no. of hours we took sleep on particular day of week
eat=[2,2,2,2,2,2,2]
code=[3,4,3,4,3,4,3]
otherwork=[5,5,5,5,5,6,5]


waste=[]                       # we will calculate the no. of hrs wasted per day and append it to the waste list
for i in range(0,7): 
    w=int(24-(sleep[i]+eat[i]+code[i]+otherwork[i]))
    waste.append(w)
    
print waste

#stackplot takes first plot parameter on x-axis and rest on y-axis
#we can use m for megenta color, c for cyan, b for blue, r for red and k for black

plt.stackplot(daysofweek, sleep,eat,code,otherwork,waste, colors=['m','c','b','r','k'], labels=['sleep','eat','code','otherwork','waste'])

plt.title("data 2018")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.legend()
plt.show()
```

```python
# 5. Now let us draw a pie chart 


import matplotlib.pyplot as plt

daysofweek=[1,2,3,4,5,6,7]

sleep=[7,8,6,7,7,8,9]  #these are no. of hours we took sleep on particular day of week
eat=[2,2,2,2,2,2,2]
code=[3,4,3,4,3,4,3]
otherwork=[5,5,5,5,5,6,5]
waste=[]                       # we will calculate the no. of hrs wasted per day and append it to the waste list
for i in range(0,7): 
    w=int(24-(sleep[i]+eat[i]+code[i]+otherwork[i]))
    waste.append(w)

day=[]      #we use day[] list to take day wise input

for i in range(0,7):
    d=[sleep[i],eat[i],code[i],otherwork[i],waste[i]]
    day.append(d)

activities=["sleep","eat","code","otherwork","waste"]    #to print labels
cols=['c','m','y','r','k']                               #list of colors

plt.pie(day[0],labels=activities,colors=cols)   #plotting pie-chart for first day i.e. day[0]

plt.title("data 2018")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.legend()
plt.show()
```
